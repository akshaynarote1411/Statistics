#### Varience

```python
np.var(l)
```

```python
# steps to calculate varience

# var = (summation(x-mean)^2)/total no. of values(N)

# 1. raw data (x)
# 2. mean(mu)
# 3. x-mu
# 4. (x-mu)^2
# 5. summation((x-mu)^2)
# 6. summation((x-mu)^2)/total no. values(N)
```

```python
l = [10,20,50,120,130,5,150,200] 
arr = np.array(l)                # 1
mean = np.mean(l)
dev = arr - mean 
sq = np.square(dev)
summ = np.sum(sq)
var = summ/len(l)

var
```

```python
np.sum(np.square(arr-mean))/len(l)
```

```python
np.sum(np.square(np.array(l)-np.mean(l)))/len(l)
```

#### Covarience

```python
np.cov(l)
```

```python
l1 = [10,20,50,120,130,5,150,200]
l2 = [2,3,5,9,10,30,40,50]

arr1 = np.array(l1)
arr2 = np.array(l2)
```

```python
# Covarience = sum((x-x_mean)(y-y_mean))/(n-1)

cov = np.sum((arr1-np.mean(arr1))*(arr2-np.mean(arr2)))/(len(l1)-1)
cov
```

```python
np.cov(arr1,arr2)
```
