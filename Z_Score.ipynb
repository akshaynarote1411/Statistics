```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df  = pd.DataFrame({"x":[30,42,47,52,55,57,63,67,68,60,70,77,79,83,88,100]})
df
```

```python
df["x"].skew()
```

```python
df["z_score"] = (df["x"] - df["x"].mean())/df["x"].std(ddof=0)     # z-score = (xi - x.mean)/ std.dev
df
```

```python
from scipy.stats import zscore
zscore(df["x"])
```

```python
from scipy.stats import norm
norm.cdf(df["x"])      
```

```python
sns.histplot(df["x"],bins=7)
```

```python
sns.histplot(df["x"],bins=7,stat="probability")
```

```python
sns.kdeplot(df["x"])
```

```python
sns.kdeplot(df["z_score"])    # When we apply z-score then for that z-score column(z_score) the mean = 0, & std. deviation = 1. It is fix for all data.
```
